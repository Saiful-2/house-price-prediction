## 🟩 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)

sns.set_style("whitegrid")
```

## 🟩 2. Load Cleaned Dataset

```python
url = "https://raw.githubusercontent.com/Saiful-2/datasets/main/ames_housing_cleaned.csv"

df = pd.read_csv(url)

df.head()
```

## 🟩 3. Overview of Dataset

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

## 4. Target Variable Analysis (VERY IMPORTANT)
📊 Distribution of SalePrice

```python
plt.figure(figsize=(8,5))
sns.histplot(df["SalePrice"], kde=True)
plt.title("Distribution of SalePrice")
plt.show()
```

### Insight:
Typically right-skewed distribution
High-priced houses are fewer

### 📊 Log Transformation (Preview for modeling)

```python
plt.figure(figsize=(8,5))
sns.histplot(np.log1p(df["SalePrice"]), kde=True)
plt.title("Log Transformed SalePrice")
plt.show()
```

### 💡 Insight:
Distribution becomes more normal
Helps regression models perform better

## 📈 5. Numerical Feature Analysis
🔥 Correlation Heatmap

```python
plt.figure(figsize=(12,10))
corr = df.corr(numeric_only=True)

sns.heatmap(corr, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap")
plt.show()
```

### Top Features Correlated with SalePrice

```python
corr_target = corr["SalePrice"].sort_values(ascending=False)
corr_target.head(10)
```

### Scatter Plots (Key Relationships)
GrLivArea vs SalePrice

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=df["Gr Liv Area"], y=df["SalePrice"])
plt.title("Gr Liv Area vs SalePrice")
plt.show()
```

### Insight:
Strong positive relationship

### OverallQual vs SalePrice

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df["Overall Qual"], y=df["SalePrice"])
plt.title("Overall Qual vs SalePrice")
plt.show()
```

### Insight:
Higher quality → significantly higher price

### GarageCars vs SalePrice

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df["Garage Cars"], y=df["SalePrice"])
plt.title("Garage Cars vs SalePrice")
plt.show()
```

## 🏘️ 6. Categorical Feature Analysis
📊 Neighborhood vs SalePrice

```python
plt.figure(figsize=(12,6))
top_neigh = df.groupby("Neighborhood")["SalePrice"].median().sort_values(ascending=False)

sns.barplot(x=top_neigh.index, y=top_neigh.values)
plt.xticks(rotation=90)
plt.title("Median SalePrice by Neighborhood")
plt.show()
```

## House Style

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df["House Style"], y=df["SalePrice"])
plt.xticks(rotation=45)
plt.title("House Style vs SalePrice")
plt.show()
```

## 7. Feature Distribution Analysis
📊 Important Numerical Features

```python
features = ["Gr Liv Area", "Total Bsmt SF", "Garage Area"]

df[features].hist(bins=30, figsize=(12,5))
plt.show()
```

### Insight:
Some features are skewed → need transformation later

## 8. Outlier Check

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=df["Gr Liv Area"], y=df["SalePrice"])
plt.title("Outlier Check")
plt.show()
```

## 9. Pairplot

```python
sample_df = df.sample(500, random_state=42)

sns.pairplot(sample_df[["SalePrice", "Gr Liv Area", "Overall Qual", "Garage Cars"]])
plt.show()
```
