## 🟩 1. Import Libraries

```python
import pandas as pd
import numpy as np

# Visualization (for quick checks)
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option("display.max_columns", None)
```

## 🟩 2. Load Dataset

```python
url = "http://raw.githubusercontent.com/Saiful-2/datasets/main/AmesHousing.csv"

df = pd.read_csv(url)

# Preview
df.shape, df.head()
```

## 🟩 3. Basic Information

```python
df.info()
```

```python
df.describe()
```

## 🟩 4. Check Missing Values

```python
missing = df.isnull().sum().sort_values(ascending=False)
missing = missing[missing > 0]

missing.head(20)
```

## 📊 Visualize Missing Values

```python
plt.figure(figsize=(12,6))
missing.head(20).plot(kind='bar')
plt.title("Top 20 Missing Features")
plt.show()
```

## 🟩 5. Drop Irrelevant Columns

```python
df.drop(columns=["Order", "PID"], inplace=True, errors='ignore')
```

## 🟩 6. Separate Numerical & Categorical

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns

len(num_cols), len(cat_cols)
```

## 🟩 7. Handle Missing Values (Categorical)

👉 Domain-based filling

```python
# Fill "None" where absence means no feature
none_cols = [
    "Alley", "MasVnrType", "BsmtQual", "BsmtCond", "BsmtExposure",
    "BsmtFinType1", "BsmtFinType2", "FireplaceQu", "GarageType",
    "GarageFinish", "GarageQual", "GarageCond", "PoolQC",
    "Fence", "MiscFeature"
]

for col in none_cols:
    if col in df.columns:
        df[col] = df[col].fillna("None")
```

```python
# Fill remaining categorical with mode
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

## 🟩 8. Handle Missing Values (Numerical)

```python
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
```

## 🟩 9. Verify Missing Values

```python
df.isnull().sum().sum()
```

## 🟩 10. Fix Data Types (Optional but Good Practice)

```python
# Example: Convert year columns if needed
year_cols = ["YearBuilt", "YearRemodAdd", "GarageYrBlt"]

for col in year_cols:
    if col in df.columns:
        df[col] = df[col].astype(int)
```

## 🟩 11. Remove Duplicates

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

## 🟩 12. Outlier Detection (Important)
📊 Example: GrLivArea vs SalePrice

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=df["Gr Liv Area"], y=df["SalePrice"])
plt.show()
```

```python
# Remove extreme outliers:
df = df[df["Gr Liv Area"] < 4000]
```

## 🟩 13. Reset Index

```python
df.reset_index(drop=True, inplace=True)
```
